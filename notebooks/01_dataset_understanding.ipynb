# Dataset Understanding

**Imports & Configurations**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import psutil

print(f"free memory: {psutil.virtual_memory().available / 1024 ** 3:.2f} GB")

pd.set_option("display.max_columns", 50)
pd.set_option("display.max_colwidth", None)
pd.set_option("display.float_format", "{:.2f}".format)

DATA_DIR = "../dataset"
FIGURES_DIR = "../outputs/figures"
os.makedirs(FIGURES_DIR, exist_ok=True)

SEED = 42
```

**Load and inspect every file**

```python
files = {
    "application_train": "application_train.csv",
    "application_test": "application_test.csv",
    "bureau":            "bureau.csv",
    "bureau_balance":    "bureau_balance.csv",
    "previous_application": "previous_application.csv",
    "pos_cash":          "POS_CASH_balance.csv",
    "installments":      "installments_payments.csv",
    "credit_card":       "credit_card_balance.csv"
}

col_des_df = pd.DataFrame(pd.read_csv(os.path.join(DATA_DIR, "HomeCredit_columns_description.csv"), encoding="cp1252"))
```

**Inspect Application Train Table**

```python
df = pd.DataFrame(pd.read_csv(os.path.join(DATA_DIR, files['application_train'])))
print("====== First data ======")
display(df.head())

print("====== Data Types ======")
display(df.dtypes.to_dict())

print("====== Summary of Data ======")
display(df.info())

# Types, nulls, and column descriptions
summary = pd.DataFrame({
    "column": df.columns,
    "types": df.dtypes,
    "nulls": df.isnull().sum(),
    "null_pct": (df.isnull().sum() / len(df) * 100).round(2)
}).reset_index(drop=True)

summary = summary.merge(
    col_des_df[["Row", "Description"]],
    left_on="column",
    right_on="Row",
    how="left"
).drop(columns=["Row"])

summary = (
    summary[summary["null_pct"] > 0]
    .sort_values("null_pct", ascending=False)
    .style.set_properties(
        subset=["column", "Description"],
        **{"text-align": "left"}
    )
)

display(summary)
```

TARGET value distribution

```python
target_counts = df["TARGET"].value_counts()
target_pct    = df["TARGET"].value_counts(normalize=True).mul(100).round(2)

print("Target distribution:")
print(pd.DataFrame({"Count": target_counts, "Percent (%)": target_pct}))

fig, ax = plt.subplots(figsize=(5, 4))
target_counts.plot(kind="bar", color=["steelblue", "tomato"],
                   edgecolor="black", ax=ax)
ax.set_title("Target Distribution\n(0 = Repaid, 1 = Default)")
ax.set_xlabel("TARGET")
ax.set_ylabel("Count")
ax.set_xticklabels(["0 — Repaid", "1 — Default"], rotation=0)
for p in ax.patches:
    ax.annotate(f"{p.get_height():,.0f}", 
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha="center", va="bottom", fontsize=10)
plt.tight_layout()
plt.savefig(f"{FIGURES_DIR}/target_distribution.png", dpi=150)
plt.show()
```

Missing value heat map for random 500 rows

```python
top_missing_cols = summary.data["column"].head(20).tolist()

fig, ax = plt.subplots(figsize=(12, 5))
sns.heatmap(
    df[top_missing_cols].isnull().astype(int).sample(n=500, random_state=SEED),
    cbar=False, yticklabels=False, cmap="Reds", ax=ax
)
ax.set_title("Missing Value Pattern — application_train (top 20 columns, random 500 rows)")
ax.set_xlabel("Column")
plt.xticks(rotation=45, ha="right", fontsize=8)
plt.tight_layout()
plt.savefig(f"{FIGURES_DIR}/application_train_missing_value_heatmap.png", dpi=150)
plt.show()
```
