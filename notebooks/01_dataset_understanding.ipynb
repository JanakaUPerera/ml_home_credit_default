**Imports & Configurations**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sns
import os
import psutil

print(f"free memory: {psutil.virtual_memory().available / 1024 ** 3:.2f} GB")

pd.set_option("display.max_columns", 50)
pd.set_option("display.max_colwidth", None)
pd.set_option("display.float_format", "{:.2f}".format)

DATA_DIR = "../dataset"
FIGURES_DIR = "../outputs/figures"
os.makedirs(FIGURES_DIR, exist_ok=True)

SEED = 42
```

**Load and inspect every file**

```python
files = {
    "application_train": "application_train.csv",
    "application_test": "application_test.csv",
    "bureau":            "bureau.csv",
    "bureau_balance":    "bureau_balance.csv",
    "previous_application": "previous_application.csv",
    "pos_cash":          "POS_CASH_balance.csv",
    "installments":      "installments_payments.csv",
    "credit_card":       "credit_card_balance.csv"
}

col_des_df = pd.DataFrame(pd.read_csv(os.path.join(DATA_DIR, "HomeCredit_columns_description.csv"), encoding="cp1252"))
```

**Inspect Application Train Table**

```python
df = pd.DataFrame(pd.read_csv(os.path.join(DATA_DIR, files['application_train'])))
print("====== First data ======")
display(df.head())

print("====== Data Types ======")
display(df.dtypes.to_dict())

# Types, nulls, and column descriptions
summary = pd.DataFrame({
    "column": df.columns,
    "types": df.dtypes,
    "nulls": df.isnull().sum(),
    "null_pct": (df.isnull().sum() / len(df) * 100).round(2)
}).reset_index(drop=True)

summary = summary.merge(
    col_des_df[["Row", "Description"]],
    left_on="column",
    right_on="Row",
    how="left"
).drop(columns=["Row"])

summary = (
    summary[summary["null_pct"] > 0]
    .sort_values("null_pct", ascending=False)
    .style.set_properties(
        subset=["column", "Description"],
        **{"text-align": "left"}
    )
)

display(summary)
```
